## imports

```python
import pandas as pd
```

```python
df_librarie = pd.read_excel('data/excelfiles/Urheber und Reihen.xlsx', sheet_name='Recherche', header=0)

df_blatter = pd.read_feather('data/feather/df_neuheiten_blatter.feather')
#df_buchhaus = pd.read_feather('data/feather/df_neuheiten_buchhaus.feather')
```

## Data Wrangling Urheber und Reihen

```python
df_librarie
```

```python
# add new column to df_libarie with the second element in alle reihen after the semicolon
df_librarie['Reihe'] = df_librarie['Alle Reihen'].str.split(';').str[1]
df_librarie['Auflage'] = df_librarie['Alle Reihen'].str.split(';').str[0]

df_librarie = df_librarie[['1. Urheber', 'Auflage', 'Reihe']]
```

## book analysis


```python
# autors which have new books on the market and are in the librarie
interresting_books = df_blatter[df_blatter['autor'].isin(df_librarie['1. Urheber'])]
interresting_books.head()
```

```python
# Reihen which have new books on the market and are in the librarie
df_blatter[df_blatter['reihe'].isin(df_librarie['Alle Reihen'])]
```

```python
# Reihennummer which have new books on the market and are in the librarie
df_blatter[df_blatter['reihenbandnummer'].isin(df_librarie['Reihennummer'])]
```

## export

```python
interresting_books.reset_index(drop=True, inplace=True)

interresting_books.to_excel('data/excelfiles/df_interresting_books.xlsx', index=False)
interresting_books.to_feather('data/feather/df_interresting_books.feather')
```
