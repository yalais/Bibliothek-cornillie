## imports

```python
import pandas as pd
```

```python
df_librarie = pd.read_excel('data/excelfiles/Urheber und Reihen.xlsx', sheet_name='Recherche', header=0)

#df_blatter = pd.read_feather('data/feather/df_neuheiten_blatter.feather')
df_buchhaus = pd.read_feather('data/feather/df_neuheiten_buchhaus.feather')
```

## Data Wrangling Urheber und Reihen

```python
df_librarie
```

```python
# add new column to df_libarie with the second element in alle reihen after the semicolon
df_librarie['Reihe'] = df_librarie['Alle Reihen'].str.split(';').str[1]
df_librarie['Auflage'] = df_librarie['Alle Reihen'].str.split(';').str[0]

df_librarie = df_librarie[['1. Urheber', 'Auflage', 'Reihe']]
```
