### Blatter Imhof AG

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get(blatter)

search = driver.find_element_by_id("rubrikArtikel:tableArtikel:0:artikelFluidGrid")
text = search.text
text = text.splitlines()
# title is every 7. element in the list
title = text[::7]
autor = text[1::7]
bookstyle = text[2::7]
language = text[3::7]
price = text[5::7]
scraped_data = pd.DataFrame({'title': title, 'autor': autor, 'bookstyle': bookstyle, 'language': language, 'price': price})

scraped_data.head(10)
```

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get(blatter)
title = []
autor = []
bookstyle = []
language = []
price = []

for i in range(5):
    search = driver.find_element_by_id("rubrikArtikel:tableArtikel:" + str(i) + ":artikelFluidGrid")
    text = search.text
    text = text.splitlines()
    # title is every 7. element in the list
    title.append(text[::7])
    autor.append(text[1::7])
    bookstyle.append(text[2::7])
    language.append(text[3::7])
    price.append(text[5::7])
    print('done with page ' + str(i))

scraped_data = pd.DataFrame({'title': title, 'autor': autor, 'bookstyle': bookstyle, 'language': language, 'price': price})
scraped_data.head(10)
```
